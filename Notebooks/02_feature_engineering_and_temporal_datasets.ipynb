# Temporal Checkpoint Dataset Generation

## Early-Warning Modeling Datasets

The previous sections generated the complete semester modeling dataset.

This section extends the feature engineering pipeline by creating multiple temporal datasets for early-warning prediction. Each dataset is generated using only the information that would have been available by a specific instructional checkpoint.

The checkpoint datasets include:

- Day 7
- Day 14
- Day 21
- Day 30

For each checkpoint:

- Student demographic and registration variables remain unchanged.
- VLE interactions are restricted to activity occurring on or before the checkpoint day.
- Assessment information is limited to submissions available by the checkpoint.
- Final course outcomes are retained only as the prediction target and are never used as predictor variables.

These datasets enable evaluation of how predictive performance changes as additional student activity becomes available throughout the semester.

# ============================================================
# TEMPORAL CHECKPOINT DATASETS
# ============================================================

```python
def build_temporal_dataset(checkpoint_day):
    """
    Build an early-warning modeling dataset using only information
    available on or before the specified course day.

    Parameters
    ----------
    checkpoint_day : int
        Final instructional day included in the dataset.

    Returns
    -------
    pandas.DataFrame
        Student-level temporal modeling dataset.
    """

    print(f"\nBuilding Day {checkpoint_day} dataset...")

    # ============================================================
    # 1. CREATE BASE STUDENT TABLE AND TARGET
    # ============================================================

    temporal_features = student_info.copy()

    temporal_features["target_success"] = (
        temporal_features["final_result"]
        .isin(["Pass", "Distinction"])
        .astype(int)
    )

    base_count = len(temporal_features)

    # ============================================================
    # 2. FILTER AND ENGINEER VLE FEATURES
    # ============================================================

    checkpoint_vle = student_vle.loc[
        student_vle["date"] <= checkpoint_day
    ].copy()

    engagement_features = (
        checkpoint_vle
        .groupby(student_keys, as_index=False)
        .agg(
            total_clicks=("sum_click", "sum"),
            average_clicks_per_record=("sum_click", "mean"),
            median_clicks_per_record=("sum_click", "median"),
            maximum_clicks_in_record=("sum_click", "max"),
            vle_records=("sum_click", "size"),
            active_days=("date", "nunique"),
            unique_vle_activities=("id_site", "nunique"),
            first_activity_day=("date", "min"),
            last_activity_day=("date", "max")
        )
    )

    engagement_features["clicks_per_active_day"] = (
        engagement_features["total_clicks"]
        / engagement_features["active_days"].replace(0, np.nan)
    )

    engagement_features["activity_span_days"] = (
        engagement_features["last_activity_day"]
        - engagement_features["first_activity_day"]
    )

    engagement_features["clicks_per_vle_activity"] = (
        engagement_features["total_clicks"]
        / engagement_features[
            "unique_vle_activities"
        ].replace(0, np.nan)
    )

    engagement_features = engagement_features.replace(
        [np.inf, -np.inf],
        np.nan
    )

    temporal_features = temporal_features.merge(
        engagement_features,
        on=student_keys,
        how="left",
        validate="one_to_one"
    )

    engagement_columns = [
        "total_clicks",
        "average_clicks_per_record",
        "median_clicks_per_record",
        "maximum_clicks_in_record",
        "vle_records",
        "active_days",
        "unique_vle_activities",
        "first_activity_day",
        "last_activity_day",
        "clicks_per_active_day",
        "activity_span_days",
        "clicks_per_vle_activity"
    ]

    temporal_features[engagement_columns] = (
        temporal_features[engagement_columns]
        .fillna(0)
    )

    temporal_features["log_total_clicks"] = np.log1p(
        temporal_features["total_clicks"]
    )

    # ============================================================
    # 3. FILTER AND ENGINEER ASSESSMENT FEATURES
    # ============================================================

    assessment_data = student_assessment.merge(
        assessments,
        on="id_assessment",
        how="left",
        validate="many_to_one"
    )

    checkpoint_assessments = assessment_data.loc[
        assessment_data["date_submitted"] <= checkpoint_day
    ].copy()

    assessment_features = (
        checkpoint_assessments
        .groupby(student_keys, as_index=False)
        .agg(
            assessments_completed=("score", "count"),
            average_score=("score", "mean"),
            median_score=("score", "median"),
            minimum_score=("score", "min"),
            maximum_score=("score", "max"),
            score_std=("score", "std"),
            average_weight=("weight", "mean"),
            total_weight_completed=("weight", "sum"),
            first_submission_day=("date_submitted", "min"),
            last_submission_day=("date_submitted", "max"),
            banked_assessments=("is_banked", "sum")
        )
    )

    assessment_features["assessment_span_days"] = (
        assessment_features["last_submission_day"]
        - assessment_features["first_submission_day"]
    )

    assessment_features["score_range"] = (
        assessment_features["maximum_score"]
        - assessment_features["minimum_score"]
    )

    weighted_scores = (
        checkpoint_assessments
        .assign(
            weighted_points=lambda df: (
                df["score"] * df["weight"]
            )
        )
        .groupby(student_keys, as_index=False)
        .agg(
            weighted_points=("weighted_points", "sum"),
            total_weight=("weight", "sum")
        )
    )

    weighted_scores["weighted_average_score"] = (
        weighted_scores["weighted_points"]
        / weighted_scores["total_weight"].replace(0, np.nan)
    )

    weighted_scores = weighted_scores[
        student_keys + ["weighted_average_score"]
    ]

    assessment_features = assessment_features.merge(
        weighted_scores,
        on=student_keys,
        how="left",
        validate="one_to_one"
    )

    temporal_features = temporal_features.merge(
        assessment_features,
        on=student_keys,
        how="left",
        validate="one_to_one"
    )

    assessment_columns = [
        "assessments_completed",
        "average_score",
        "median_score",
        "minimum_score",
        "maximum_score",
        "score_std",
        "average_weight",
        "total_weight_completed",
        "first_submission_day",
        "last_submission_day",
        "banked_assessments",
        "assessment_span_days",
        "score_range",
        "weighted_average_score"
    ]

    temporal_features[assessment_columns] = (
        temporal_features[assessment_columns]
        .fillna(0)
    )

    temporal_features["score_improvement"] = (
        temporal_features["maximum_score"]
        - temporal_features["average_score"]
    )

    temporal_features["assessment_intensity"] = (
        temporal_features["assessments_completed"]
        / (
            temporal_features["assessment_span_days"] + 1
        )
    )

    temporal_features["log_assessments_completed"] = np.log1p(
        temporal_features["assessments_completed"]
    )

    # ============================================================
    # 4. ENGINEER CHECKPOINT-SAFE REGISTRATION FEATURES
    # ============================================================

    registration_features = student_registration.copy()

    registration_features["registered_before_start"] = (
        registration_features["date_registration"] < 0
    ).astype(int)

    registration_features["registered_after_start"] = (
        registration_features["date_registration"] > 0
    ).astype(int)

    registration_features["days_registered_before_start"] = (
        -registration_features["date_registration"]
    ).clip(lower=0)

    # Withdrawal is known only when it occurred by the checkpoint.
    withdrawal_known = (
        registration_features["date_unregistration"].notna()
        & (
            registration_features["date_unregistration"]
            <= checkpoint_day
        )
    )

    registration_features["withdrew"] = (
        withdrawal_known.astype(int)
    )

    registration_features["days_until_withdrawal"] = np.where(
        withdrawal_known,
        registration_features["date_unregistration"]
        - registration_features["date_registration"],
        0
    )

    registration_features["early_withdrawal"] = (
        withdrawal_known
        & (
            registration_features["date_unregistration"]
            <= 60
        )
    ).astype(int)

    # Hide future withdrawal dates to prevent leakage.
    registration_features["date_unregistration"] = np.where(
        withdrawal_known,
        registration_features["date_unregistration"],
        0
    )

    registration_features["registration_timing"] = pd.cut(
        registration_features["date_registration"],
        bins=[
            -np.inf,
            -60,
            -30,
            -1,
            0,
            30,
            np.inf
        ],
        labels=[
            "very_early",
            "early",
            "moderately_early",
            "course_start",
            "late",
            "very_late"
        ]
    )

    registration_columns = student_keys + [
        "date_registration",
        "date_unregistration",
        "registered_before_start",
        "registered_after_start",
        "days_registered_before_start",
        "withdrew",
        "days_until_withdrawal",
        "early_withdrawal",
        "registration_timing"
    ]

    registration_features = registration_features[
        registration_columns
    ]

    temporal_features = temporal_features.merge(
        registration_features,
        on=student_keys,
        how="left",
        validate="one_to_one"
    )

    registration_numeric_columns = [
        "date_registration",
        "date_unregistration",
        "registered_before_start",
        "registered_after_start",
        "days_registered_before_start",
        "withdrew",
        "days_until_withdrawal",
        "early_withdrawal"
    ]

    temporal_features[registration_numeric_columns] = (
        temporal_features[registration_numeric_columns]
        .fillna(0)
    )

    temporal_features["registration_timing"] = (
        temporal_features["registration_timing"]
        .astype("object")
        .fillna("unknown")
    )

    # ============================================================
    # 5. CLEAN DEMOGRAPHIC VARIABLES
    # ============================================================

    demographic_categorical_columns = [
        "imd_band",
        "region",
        "highest_education",
        "age_band",
        "gender",
        "disability"
    ]

    temporal_features[demographic_categorical_columns] = (
        temporal_features[demographic_categorical_columns]
        .fillna("Unknown")
    )

    temporal_features["gender"] = (
        temporal_features["gender"]
        .map({
            "M": 0,
            "F": 1
        })
        .fillna(0)
        .astype(int)
    )

    temporal_features["disability"] = (
        temporal_features["disability"]
        .map({
            "N": 0,
            "Y": 1
        })
        .fillna(0)
        .astype(int)
    )

    # ============================================================
    # 6. ONE-HOT ENCODE CATEGORICAL VARIABLES
    # ============================================================

    categorical_columns = [
        "region",
        "highest_education",
        "imd_band",
        "age_band",
        "registration_timing"
    ]

    temporal_features = pd.get_dummies(
        temporal_features,
        columns=categorical_columns,
        drop_first=True,
        dtype=int
    )

    # Add checkpoint identifier.
    temporal_features["checkpoint_day"] = checkpoint_day

    # ============================================================
    # 7. VALIDATE THE DATASET
    # ============================================================

    duplicate_count = temporal_features.duplicated(
        subset=student_keys
    ).sum()

    if len(temporal_features) != base_count:
        raise ValueError(
            f"Day {checkpoint_day}: student count changed "
            f"from {base_count} to {len(temporal_features)}."
        )

    if duplicate_count != 0:
        raise ValueError(
            f"Day {checkpoint_day}: "
            f"{duplicate_count} duplicate records found."
        )

    print(
        f"Day {checkpoint_day} complete: "
        f"{temporal_features.shape[0]} rows, "
        f"{temporal_features.shape[1]} columns"
    )

    return temporal_features
```

```python
DAY_7 = build_temporal_dataset(checkpoint_day=7)
DAY_14 = build_temporal_dataset(checkpoint_day=14)
DAY_21 = build_temporal_dataset(checkpoint_day=21)
DAY_30 = build_temporal_dataset(checkpoint_day=30)
```

```python
checkpoint_datasets = {
    7: DAY_7,
    14: DAY_14,
    21: DAY_21,
    30: DAY_30
}

for checkpoint_day, dataset in checkpoint_datasets.items():
    print(
        f"Day {checkpoint_day}: "
        f"{dataset.shape[0]} rows, "
        f"{dataset.shape[1]} columns, "
        f"checkpoint values = "
        f"{dataset['checkpoint_day'].unique()}"
    )
```

```python
output_files = {
    7: "early_warning_day7_dataset.csv",
    14: "early_warning_day14_dataset.csv",
    21: "early_warning_day21_dataset.csv",
    30: "early_warning_day30_dataset.csv"
}

for checkpoint_day, dataset in checkpoint_datasets.items():
    filename = output_files[checkpoint_day]

    dataset.to_csv(
        filename,
        index=False
    )

    print(
        f"Saved Day {checkpoint_day}: "
        f"{filename} "
        f"({dataset.shape[0]} rows, "
        f"{dataset.shape[1]} columns)"
    )
```
